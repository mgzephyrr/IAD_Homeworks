Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?

```python
passmark = 50
```

### Код

```python
passed = len(df.loc[df['math score'] >= passmark]) / len(df)
prep_passed = len(df.loc[(df['math score'] >= passmark) & (df['test preparation'] == 'completed')]) / len(df.loc[df['test preparation'] == 'completed'])
woman_no_prep_failed = len(df.loc[(df['math score'] < passmark) & (df['test preparation'] == 'none') & (df['gender'] == 'female')]) / len(df.loc[(df['test preparation'] == 'none') & (df['gender'] == 'female')])
```

### Результат работы

```python
print('Сдали:', passed, '\nПодготовились и сдали', prep_passed, '\nЖенщины, которые не готовились и не сдали:', woman_no_prep_failed)
```

## Задание 16

### Формулировка

Переведем баллы в оценки

Система перевода баллов в оценки

больше 90 = A

80-90 = B

70-80 = C

60-70 = D

50-60 = E

меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).

### Код

```python
def GetGrade(average_mark):
    if average_mark > 90:
        return 'A'
    if 80 <= average_mark <= 90:
        return 'B'
    if 70 <= average_mark < 80:
        return 'C'
    if 60 <= average_mark < 70:
        return 'D'
    if 50 <= average_mark < 60:
        return 'E'
    return 'F'

df['Grade'] = df.apply(lambda row: GetGrade((row['math score'] + row['reading score'] + row['writing score']) / 3), axis=1)
df
```

```python
df['Grade'].value_counts()
```
